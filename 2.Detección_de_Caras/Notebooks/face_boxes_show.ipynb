```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import json

with open("interp-test-out.json", "r") as read_file:
    data = json.load(read_file)

first_face_file =  list(data.keys())[0]
second_face_file =  list(data.keys())[1]
third_face_file =  list(data.keys())[2]

first_face_points = data[first_face_file][0]
print("First face points {}".format(first_face_points))
second_face_points = data[second_face_file][0]
print("Second face points {}".format(second_face_points))
third_face_points = data[third_face_file][0]
print("Third face points {}".format(third_face_points))

first_img = cv2.imread(first_face_file)
second_img = cv2.imread(second_face_file)
third_img = cv2.imread(third_face_file)

(x,y,w,h) = first_face_points
cv2.rectangle(first_img, (x, y), (w, h), (255, 0, 0), 2)
plt.imshow(first_img)
plt.show()

(x,y,w,h) = second_face_points
cv2.rectangle(second_img, (x, y), (w, h), (255, 0, 0), 2)
plt.imshow(second_img)
plt.show()

(x,y,w,h) = third_face_points
cv2.rectangle(third_img, (x, y), (w, h), (255, 0, 0), 2)
plt.imshow(third_img)
plt.show()
```
