```python
import lstm_test as lt
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
precisions_and_recalls = lt.generate_outcomes()

thresholds = []
precisions = []
recalls = []
data = []

for key, value in precisions_and_recalls.items():
    thresholds.append(key)
    aux_prec = []
    aux_rec = []
    aux_data = []
    for element in value:

        aux_prec.append(element[0])

        aux_rec.append(element[1])

        aux_data.append(element[3])

    precisions.append(aux_prec)
    recalls.append(aux_rec)
    data.append(aux_data)

precisions = np.asarray(precisions)
recalls = np.asarray(recalls)
precisions = [['%.4f' % j for j in i] for i in precisions]
recalls = [['%.4f' % j for j in i] for i in recalls]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
table = ax.table(cellText=precisions, 
                rowLabels=thresholds,
                colLabels=range(4),
                loc='center')
fig.suptitle('PrecisiÃ³n',fontsize=32)
table.auto_set_font_size(False)
table.set_fontsize(24)
table.scale(4, 4)
ax.axis('off')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
table = ax.table(cellText=recalls, 
                rowLabels=thresholds,
                colLabels=range(4),
                loc='center')
fig.suptitle('Exhaustividad',fontsize=32)
table.auto_set_font_size(False)
table.set_fontsize(24)
table.scale(4, 4)
ax.axis('off')
plt.show()
```

```python
for idx,value in enumerate(data):
    title = thresholds [idx]
    for jdx,element in enumerate(value):
        print(str(title) + ' ' + str(jdx))
        data_pro = np.asarray(list(element.values())).reshape((2,2),order='F')
        fig = plt.figure()
        fig.suptitle('Threshold = ' + str(title) +', ' + 'criterio = ' + str(jdx), fontsize=24)
        ax = fig.add_subplot(1,1,1)
        table = ax.table(cellText=data_pro, 
                        rowLabels=['Real Positivo', 'Real Negativo'],
                        colLabels=['Predecido Positivo','Predecido Negativo'],
                        loc='center')
        
        table.auto_set_font_size(False)
        table.set_fontsize(24)
        table.scale(4, 4)
        ax.axis('off')
        plt.show()
        
```
